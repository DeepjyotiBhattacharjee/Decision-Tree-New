```python
from sklearn.datasets import load_iris
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.model_selection import train_test_split
```

```python
iris = load_iris()
iris
```

```python
y = iris.target
y
```

```python
X = iris.data
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_true=y_test,y_pred = y_pred)
```

```python
# we are getting 94.7 percent accuracy
```

### Hyperparameter tuning

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'max_depth' :[2,3,4,5,6],
    'min_samples_split' : [2,3,4,5],
    'min_samples_leaf' : [2,3,4]
}
```

```python
grid_search = GridSearchCV(estimator=model,param_grid=param_grid,cv=5)
```

```python
grid_search.fit(X_train,y_train)
```

```python
grid_search.best_estimator_
```

```python
grid_search.best_params_
```

```python
# now that we have the best params, let's create another model with these params
```

```python
model2 = DecisionTreeClassifier(max_depth= 3, min_samples_leaf= 2, min_samples_split= 2)
```

```python
model2.fit(X_train,y_train)
```

```python
y_pred2 = model2.predict(X_test)
y_pred2
```

```python
accuracy_score(y_true=y_test,y_pred=y_pred2)
```

```python
model.score(X_train,y_train)
```

```python
model2.score(X_train,y_train)
```

```python
# Note that the difference between training and testing accuracy should not be more than 5 percent
```
